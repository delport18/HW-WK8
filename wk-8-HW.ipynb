````python
dfo_topg = dfo.query("Medal == 'Gold'")["NOC"].value_counts().to_frame("Quantity").sort_values(by = "Quantity", ascending = False)
dfo_topg.head(10)
````

````python
## bring in our diamonds dataset
dfd = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/refs/heads/main/diamonds.csv")
dfd
````

````python
dfd["diamond cut"].nunique()
````

````python
dfd["diamond cut"].value_counts(ascending = False)
````

````python
dfd["diamond cut"].value_counts().to_frame("Quantity").sort_values(by = "Quantity", ascending = True)
````

````python
## Normailze gives you the percentages
dfd["diamond cut"].value_counts(normalize = True)
````

````python
round(dfd["diamond cut"].value_counts(normalize = True),2)
````

````python
round(dfd["diamond cut"].value_counts(normalize = True) *100 ,1)
````
